```python
import pandas as pd
import numpy as np
```

```python
total_score=deliveries.groupby(['id','inning']).sum()['total_runs'].reset_index()
total_score
```

```python
total_score=total_score[total_score['inning']==1]
total_score['total_runs']=total_score['total_runs']+1
```

```python
match_df=match.merge(total_score[['id','total_runs']],left_on='id',right_on='id')
match_df
```

```python
current_teams=[
    'Royal Challengers Bangalore',
    'Kings XI Punjab',
    'Delhi Capitals',
    'Mumbai Indians',
    'Kolkata Knight Riders',
    'Rajasthan Royals',
    'Chennai Super Kings',
    'Sunrisers Hyderabad'
]
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df['team1'].unique()
```

```python
match_df=match_df[match_df['team1'].isin(current_teams)]
match_df=match_df[match_df['team2'].isin(current_teams)]
```

```python
match_df['team1'].unique()
```

```python
match_df.shape
```

```python
#match_df['method'].value_counts()
match_df=match_df[match_df['method']!='D/L']
```

```python
match_df=match_df[['id','city','winner','total_runs']]
match_df
```

```python
deliveries_df=match_df.merge(deliveries,on='id')
```

```python
deliveries_df=deliveries_df[deliveries_df['inning']==2]
```

```python
deliveries_df.shape
```

```python
deliveries_df['current score']=deliveries_df.groupby('id').cumsum()['total_runs_y']
```

```python
deliveries_df['runs left']=deliveries_df['total_runs_x']-deliveries_df['current score']
```

```python
deliveries_df['ball left']=120-(deliveries_df['over']*6+deliveries_df['ball'])
```

```python
deliveries_df
```

```python
deliveries_df['player_dismissed']=deliveries_df['player_dismissed'].fillna('0')
deliveries_df['player_dismissed']=deliveries_df['player_dismissed'].apply(lambda x:x if x=='0' else '1' )
deliveries_df['player_dismissed']=deliveries_df['player_dismissed'].astype('int')
wickets=deliveries_df.groupby('id').cumsum()['player_dismissed'].values
deliveries_df['wickets']=10-wickets
```

```python
deliveries_df
```

```python
#deliveries_df[deliveries_df['ball left']==120]
#deliveries_df[deliveries_df['over']==19]
#import seaborn as sns
#sns.countplot(deliveries_df['over'])

```

```python
#current run rate
deliveries_df['CRR']=(deliveries_df['current score']*6)/(120-deliveries_df['ball left'])
```

```python
#required run rate
deliveries_df['RRR']=(deliveries_df['runs left']*6)/(deliveries_df['ball left'])
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
deliveries_df['result']=deliveries_df.apply(result,axis=1)
```

```python
deliveries_df
```

```python
final_df=deliveries_df[['batting_team','bowling_team','city','runs left','ball left','wickets','total_runs_x','CRR','RRR','result']]
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df['city'].unique()
```

```python
final_df.dropna(inplace=True)# we can also fill that value
```

```python
final_df=final_df[final_df['ball left']!=0]
```

```python
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
